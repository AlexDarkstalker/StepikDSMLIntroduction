```python
import pandas as pd
import numpy as np
import seaborn as sns
```

```python
alien_data = pd.read_csv("alien_invasion/invasion.csv")
alien_data
```

```python
alien_data.isna().value_counts()
```

```python
X_train = alien_data.drop(["class"], axis=1)
X_train
```

```python
y_train = alien_data['class']
```

```python
y_train
```

```python
from sklearn.ensemble import RandomForestClassifier
from sklearn.model_selection import GridSearchCV
```

```python
rf = RandomForestClassifier(random_state=0)
```

```python
params = {"n_estimators": range(10, 51, 10), "max_depth": range(1, 13, 2), \
          "min_samples_leaf": range(1, 8), "min_samples_split": range(2, 10, 2)}
```

```python
search = GridSearchCV(rf, param_grid=params, cv=3, n_jobs=-1)
```

```python
search.fit(X_train, y_train)
```

```python
search.best_params_
```

```python
X_test = pd.read_csv("alien_invasion/operative_information.csv")
X_test
```

```python
prediction = search.predict(X_test)
prediction
```

```python
pd.Series(prediction).value_counts()
```

```python
pd.DataFrame(search.best_estimator_.feature_importances_, index=X_test.columns).sort_values(by=0, ascending=False)
```

```python
space_data = pd.read_csv("alien_invasion/space_can_be_a_dangerous_place.csv")
space_data
```

```python
space_data.isna().value_counts()
```

```python
X_train_space = space_data.drop("dangerous", axis=1)
X_train_space
```

```python
y_train_space = space_data['dangerous']
```

```python
space_rf = RandomForestClassifier(random_state=0)
```

```python
space_search = GridSearchCV(space_rf, param_grid=params, cv=3, n_jobs=-1)
```

```python
space_search.fit(X_train_space, y_train_space)
```

```python
pd.DataFrame(space_search.best_estimator_.feature_importances_, index=X_train_space.columns, columns=["importance"])\
                        .sort_values(by="importance", ascending=False)
```
