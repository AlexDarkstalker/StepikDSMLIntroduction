```python
max_depth_range = range(1, 100)
```

```python
from sklearn.metrics import accuracy_score
from sklearn.model_selection import cross_val_score
```

```python
np.random.seed(0)
```

```python
iris_scores = pd.DataFrame(columns=['max_depth', 'train_score', 'test_score'])
for max_depth in max_depth_range:
    clf = tree.DecisionTreeClassifier(criterion='entropy', max_depth=max_depth)
    clf.fit(X_train, y_train)
    train_score = clf.score(X_train, y_train)
    y_pred = clf.predict(X_test)
    test_score = accuracy_score(y_true=y_test, y_pred=y_pred)
    cv_score = cross_val_score(clf, X_train, y_train, cv=5).mean()
#     test_score = clf.predict(X_test, y_test)
#     test_score = clf.score(X_test, y_test)
    tmp_score = pd.DataFrame({
        'max_depth': [max_depth],
        'train_score': [train_score],
        'test_score': [test_score],
        'mean_cv_score': [cv_score]
                })
    iris_scores = iris_scores.append(tmp_score)
iris_scores
    
```

```python
iris_scores_long = iris_scores.melt(id_vars=['max_depth'], value_vars=['train_score', 'test_score', 'mean_cv_score'],
                                   var_name='set_type', value_name='scores')
```

```python
iris_scores_long
```

```python
sns.lineplot(data=iris_scores_long, x='max_depth', y='scores', hue='set_type')
```
