```python
students_pd[students_pd.lunch == 'free/reduced'].shape[0] / students_pd.shape[0]
```

```python
reduced_students = students_pd[students_pd.lunch == 'free/reduced']
```

```python
standard_students = students_pd[students_pd.lunch == 'standard']
```

```python
re_math_mean = reduced_students['math score'].mean()
re_read_mean = reduced_students['reading score'].mean()
re_wr_mean = reduced_students['writing score'].mean()
re_math_var = reduced_students['math score'].var()
re_read_var = reduced_students['reading score'].var()
re_wr_var = reduced_students['writing score'].var()
```

```python
import math
math.sqrt(re_math_var)
```

```python
st_math_mean = standard_students['math score'].mean()
st_read_mean = standard_students['reading score'].mean()
st_wr_mean = standard_students['writing score'].mean()
st_math_var = standard_students['math score'].var()
st_read_var = standard_students['reading score'].var()
st_wr_var = standard_students['writing score'].var()
```

```python
mean_series = pd.Series([re_math_mean, re_read_mean, re_wr_mean, st_math_mean, st_read_mean, st_wr_mean], 
                        index=['re_math_mean', 're_read_mean', 're_wr_mean', 'st_math_mean', 'st_read_mean', 'st_wr_mean'])
mean_series
```

```python
var_series = pd.Series([math.sqrt(re_math_var), math.sqrt(re_read_var), math.sqrt(re_wr_var), 
                        math.sqrt(st_math_var), math.sqrt(st_read_var), math.sqrt(st_wr_var)],
                      index=['re_math_var', 're_read_var', 're_wr_var', 'st_math_var', 'st_read_var', 'st_wr_var'])
var_series
```

```python
my_stat['V5'] = my_stat['V1'] + my_stat['V4']
my_stat['V6'] = np.log(my_stat.V2)
```

```python
my_stat.rename(columns={'V1': 'session_value', 'V2': 'group', 'V3': 'time', 'V4': 'n_users'}, inplace=True)
```

```python
my_stat
```

```python
my_stat.fillna(value={'session_value': 0}, inplace=True)
```

```python
my_stat.loc[my_stat.n_users < 0, 'n_users'] = my_stat[my_stat.n_users > 0]['n_users'].median()
```

```python
mean_session_value_data = my_stat.groupby('group', as_index=False).agg({'session_value': 'mean'}).rename(columns={'session_value': 'mean_session_value'})
mean_session_value_data
```
